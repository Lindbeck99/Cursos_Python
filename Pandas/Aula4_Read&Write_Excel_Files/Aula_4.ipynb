```python
import pandas
arquivo_csv = pandas.read_csv('/content/stock_data.csv', na_values = ['not available', 'n.a.'])
arquivo_csv
```

```python
arquivo_csv.to_csv('New Stock_Data')
```

```python
excel = pandas.read_excel('/content/stock_data.xlsx', nrows=3)
excel_stock = excel[['tickers', 'people']]
excel_stock
```

```python
excel_wheater = pandas.read_excel('/content/stocks_weather.xlsx', nrows=3)
excel_wheater = excel_wheater[['price', 'tickers']]
excel_wheater
```

```python
with pandas.ExcelWriter('/content/stocks_weather.xlsx') as writer:
  excel_stock.to_excel(writer, sheet_name = 'stocks')
  excel_wheater.to_excel(writer, sheet_name = "wheater")
```
