```python
import pandas
clima_pr = pandas.DataFrame({
    'Cidades': ['Curitiba', 'Guaratuba', 'Ponta Grossa'],  
    'Clima': [22, 26, 20],
    'Humidade': [67, 85, 58]  
})
clima_pr
```

```python
clima_rs = pandas.DataFrame({
    'Cidades': ['Caxias', 'Porto Alegre', 'Santa Maria'],  
    'Clima': [13, 19, 15],
    'Humidade': [71, 80, 80]  
})
clima_rs
```

```python
clima = pandas.concat([clima_rs, clima_pr], ignore_index=True)
clima
```

```python
clima = pandas.concat([clima_rs, clima_pr], keys=['Rio Grande do Sul', 'Paraná'])
clima
```
